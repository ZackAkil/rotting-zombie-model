# Generate Zombie data
We are going to generate 2 data sets for binary classification.

The first data set is what we will use to train our intial model, and to test the model intially after it's deployed.

The second data set is going to be used to simulate the zombie population mutating after the intial model is deployed. 

```
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from sklearn.externals import joblib
```

## Generate first data set

```
x_1, y_1 = make_classification(random_state=99, n_features=4, n_redundant=1)
plt.scatter(x_1[:,0], x_1[:,1], c=y_1)
plt.show()
x_1.shape, y_1.shape
```

```
# save data
joblib.dump({"x":x_1, "y":y_1}, "zombie_data_1.joblib" )
```

## Generate second data set

```
x_2, y_2 = make_classification(random_state=16, n_features=4, n_redundant=1)
plt.scatter(x_2[:,0], x_2[:,1], c=y_1)
plt.show()
x_2.shape, y_2.shape
```

```
# save data
joblib.dump({"x":x_2, "y":y_2}, "zombie_data_2.joblib" )
```
